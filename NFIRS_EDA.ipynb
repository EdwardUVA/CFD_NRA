```python
plt.figure(figsize=(20,12))
sns.set(style="ticks")

sns.histplot(data=df, x='Incident Date', hue='Incident Primary Action Taken', shrink=.8).set(title = "Charlottsville Fire Data") #bins = years, hue="FireType", 

plt.show
#plt.savefig('CFD_Year_Plot.png')
```

```python
# df['date2'] = pd.to_datetime(df['Incident Date'],format='%Y%m%d')
df['year'] = pd.DatetimeIndex(df['Incident Date']).year
df['month'] = pd.DatetimeIndex(df['Incident Date']).month
```

```python
ems_list = ['Provide first aid & check for injuries', 'Provide basic life support (BLS)', 'Provide advanced life support (ALS)', 'Transport person', 'Emergency medical services, other', 'Assistance, other'] 
ems_list
```

```python
df.loc[df['Incident Primary Action Taken'] == 'Provide first aid & check for injuries']
```

```python
#this is what I wanted.  https://stackoverflow.com/questions/12096252/use-a-list-of-values-to-select-rows-from-a-pandas-dataframe
# evertying in our ems list

ems_df = df[df['Incident Primary Action Taken'].isin(ems_list)]

ems_df
```

```python
plt.figure(figsize=(20,12))
sns.set(style="ticks")

#bin 54 = 4.5 years

sns.histplot(data=ems_df, x='Incident Date', hue='Incident Primary Action Taken', stat='count', bins = 54, shrink=.8).set(title = "Charlottesville Fire/EMS by Type, binned by year and month") #, hue="FireType", 

plt.show
```

```python
plt.figure(figsize=(20,12))
sns.set(style="ticks")
sns.color_palette("bright")

sns.histplot(data=ems_df, x='year', hue='Incident Primary Action Taken', shrink=0.8, bins = 5).set(title = "Charlottesville Fire/EMS by Type by Year") #bins = years, hue="FireType", 

plt.show
#plt.savefig('CFD_Year_Plot.png')a
```

```python
plt.figure(figsize=(20,12))
sns.set(style="ticks")

#bin 54 = 4.5 years

sns.histplot(data=ems_df, x='Incident Date', hue='Basic Shift Or Platoon', stat='count', bins = 54, shrink=.8).set(title = "Charlottesville Fire/EMS by Shift") #, hue="FireType", 

plt.show
```
