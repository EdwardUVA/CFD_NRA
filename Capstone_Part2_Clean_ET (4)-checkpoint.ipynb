## Rename UVA Neighborhoods

As requested by CFD, now that we have our "big table" lets roll up our data by neigherborhood and clean up our UVA locations

Here are the groupings I’d recommend:

UVA Main Grounds
* UVA Main Grounds
* UVA Piedmont Faculty
* UVA Fontaine
* UVA International Residential College
* UVA Lambeth
 
UVA Medical Grounds
* UVA Old Medical Grounds
* UVA

UVA North Grounds
* UVA Darden/Law/JPJ
* UVA Ivy
* UVA Carruthers Hall

We'll do what we did last time with an if statement and an apply

```python
# take neigherborhood name and change it to new UVA name as applicable

def new_nhood(name):
    if name == 'UVA Main Grounds':
        return 'UVA Main Grounds'
    elif name == 'UVA Piedmont Faculty':
        return 'UVA Main Grounds'
    elif name == 'UVA Fontaine':
        return 'UVA Main Grounds'
    elif name == 'UVA International Residential College':
        return 'UVA Main Grounds'
    elif name == 'UVA Lambeth':
        return 'UVA Main Grounds'
    elif name == 'UVA Old Medical Grounds':
        return 'UVA Medical Grounds'
    elif name  == 'UVA':
        return 'UVA Medical Grounds'
    elif name  == 'UVA Darden/Law/JPJ':
        return 'UVA North Grounds'
    elif name  == 'UVA Ivy':
        return 'UVA North Grounds'
    elif name  == 'UVA Darden/Law/JPJ':
        return 'UVA Carruthers Hall'
    else:
        return name
```

```python
df_big['New Neighborhood'] = df_big['Neighborhood'].apply(new_nhood)
```

## Ranked EMS Results by Neighborhood

Using our EMS data from the "big table" lets look at a few other ways to consider our incidnents.  Lets bring in neighborhod data for population and area.  Note that UVA doesn not have population data.

```python
hoods = pd.read_csv('neighborhoods.csv')
hoods.rename(columns = {'Neighborhood': 'New Neighborhood'}, inplace = True)
hoods
```

Now lets pull out the top 20 calls from each neighborhood.

```python
EMS_nhood[EMS_nhood['New Neighborhood'] == 'Belmont'].sort_values(by = 'EMS Total Time by Incident', ascending = False)['Incident Complaint Reported By Dispatch'].value_counts()[0:20].reset_index()
```

```python
EMS_nhood
```

```python
top20_nhood_per = pd.DataFrame()
for n in nhoods:
    temp_df = EMS_nhood[EMS_nhood['New Neighborhood'] == n].sort_values(by = 'EMS_personel_by_incident', ascending = False)[['Incident Complaint Reported By Dispatch', 'New Neighborhood']].value_counts().to_frame().reset_index()
    temp_df.rename(columns = {0: 'Count of Incindent'}, inplace = True)
    top20_nhood_per = top20_nhood_per.append(temp_df, ignore_index=True)
```

```python
top20_nhood_per
```

```python
top20_hoods = top20_nhood_per.merge(hoods, how = 'left', on = 'New Neighborhood')
top20_hoods
```

```python
top20_hoods['Incident per day pop'] = round(top20_hoods['Count of Incindent'] / top20_hoods['Population (Day)'], 4)
top20_hoods['Incident per night pop'] = round(top20_hoods['Count of Incindent'] / top20_hoods['Population (night)'], 4)
top20_hoods['Incident per area'] = round(top20_hoods['Count of Incindent'] / top20_hoods['Area (mi sq)'], 4)

top20_hoods
```

```python
top5_ems_per_nhood = pd.DataFrame()
for n in nhoods:
    temp_df = top20_hoods[top20_hoods['New Neighborhood'] == n].sort_values(by = 'Count of Incindent', ascending = False)[['Incident Complaint Reported By Dispatch', 'New Neighborhood', 'Count of Incindent']]#.value_counts()[0:5].to_frame().reset_index()
    top5_ems_per_nhood = top5_ems_per_nhood.append(temp_df, ignore_index=True)
top5_ems_per_nhood
```

```python
top5_ems_per_nhood_day = pd.DataFrame()
for n in nhoods:
    temp_df = top20_hoods[top20_hoods['New Neighborhood'] == n].sort_values(by = 'Incident per day pop', ascending = False)[['Incident Complaint Reported By Dispatch', 'New Neighborhood', 'Count of Incindent', 'Incident per day pop']]#.value_counts()[0:5].to_frame().reset_index()
    top5_ems_per_nhood_day = top5_ems_per_nhood_day.append(temp_df, ignore_index=True)
top5_ems_per_nhood_day
```

```python
top5_ems_per_nhood_night = pd.DataFrame()
for n in nhoods:
    temp_df = top20_hoods[top20_hoods['New Neighborhood'] == n].sort_values(by = 'Incident per night pop', ascending = False)[['Incident Complaint Reported By Dispatch', 'New Neighborhood', 'Count of Incindent', 'Incident per night pop']]#.value_counts()[0:5].to_frame().reset_index()
    top5_ems_per_nhood_night = top5_ems_per_nhood_night.append(temp_df, ignore_index=True)
top5_ems_per_nhood_night
```

```python
top5_ems_per_nhood_area = pd.DataFrame()
for n in nhoods:
    temp_df = top20_hoods[top20_hoods['New Neighborhood'] == n].sort_values(by = 'Incident per area', ascending = False)[['Incident Complaint Reported By Dispatch', 'New Neighborhood', 'Count of Incindent', 'Incident per area']]#.value_counts()[0:5].to_frame().reset_index()
    top5_ems_per_nhood_area = top5_ems_per_nhood_area.append(temp_df, ignore_index=True)
top5_ems_per_nhood_area
```
